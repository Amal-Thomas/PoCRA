```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the first CSV file into a pandas dataframe
deep = pd.read_csv('output/deep_daily_soil_water_balance.csv')

total_runoff_deep = deep.iloc[:, 2]
uptake_deep = deep.iloc[:, 3]
sm_deep = deep.iloc[:, 4]
gw_deep = deep.iloc[:, 5]

# Read the second CSV file into a pandas dataframe
shallow = pd.read_csv('output/shallow_daily_soil_water_balance.csv')

total_runoff_shallow = shallow.iloc[:, 2]
uptake_shallow = shallow.iloc[:, 3]
sm_shallow = shallow.iloc[:, 4]
gw_shallow = shallow.iloc[:, 5]
```

```python
# Plotting Total runoff vs days
plt.plot(total_runoff_deep, label='Deep')
plt.plot(total_runoff_shallow, label='Shallow')
plt.xlabel('Day')
plt.ylabel('Total runoff in mm')
plt.title('Total runoff')
plt.legend()
plt.show()
```

```python
# Plotting Crop water uptake vs days
plt.plot(uptake_deep, label='Deep')
plt.plot(uptake_shallow, label='Shallow')
plt.xlabel('Day')
plt.ylabel('Crop water uptake in mm')
plt.title('Crop water uptake')
plt.legend()
plt.show()
```

```python
# Plotting Soil moisture vs days
plt.plot(sm_deep, label='Deep')
plt.plot(sm_shallow, label='Shallow')
plt.xlabel('Day')
plt.ylabel('Soil moisture in mm')
plt.title('Soil moisture')
plt.legend()
plt.show()
```

```python
# Plotting Percolation to groundwater vs days
plt.plot(gw_deep, label='Deep')
plt.plot(gw_shallow, label='Shallow')
plt.xlabel('Day')
plt.ylabel('Percolation to groundwater in mm')
plt.title('Percolation to groundwater')
plt.legend()
plt.show()
```

```python
# Analysis on Crop water uptake
demand = 4
print(f'For the deep soil type, the soil moisture demand for the plant was satisfied for {sum(uptake_deep == demand)} days')
print(f'For the shallow soil type, the soil moisture demand for the plant was satisfied for {sum(uptake_shallow == demand)} days')
```

```python
# Analysis on Percolation to groundwater
print(f'For the deep soil type, the groundwater was recharged by {round(sum(gw_deep), 3)} mm')
print(f'For the shallow soil type, the groundwater was recharged by {round(sum(gw_shallow), 3)} mm')
```

```python
# Analysis on Total Runoff
print(f'For the deep soil type, the total runoff for all days in the simulation was {round(sum(total_runoff_deep), 3)} mm')
print(f'For the shallow soil type, the total runoff for all days in the simulation was {round(sum(total_runoff_shallow), 3)} mm')
```
